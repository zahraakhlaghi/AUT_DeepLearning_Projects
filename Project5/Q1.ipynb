```python
import torch
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/dataset/data.csv')
data = data.drop(['Unnamed: 0'], axis=1)
data
```

## Data Windowing

```python
import numpy as np


def data_windowing(seq,ws):
    x = []
    y = []
    L = len(seq)

    for i in range(L-ws):

        x.append(data.iloc[i:i+ws, :-1])  # Exclude the label column
        y.append(data.iloc[i+ws, -1])

    return np.array(x) , np.array(y)
```

```python
window_size = 10

x_train, y_train = data_windowing(train_data,window_size)
x_test, y_test = data_windowing(test_data,window_size)
```

```python
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return torch.tensor(self.X[idx]), torch.tensor(self.y[idx])
```

```python
from torch.utils.data import DataLoader

dataset = TimeSeriesDataset(x_train, y_train)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)  # Adjust batch size

dataset = TimeSeriesDataset(x_test, y_test)
test_dataloader = DataLoader(dataset, batch_size=32, shuffle=False)  # Adjust batch size
```

## Train RRN Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=2, dropout=0.2):
        super(RNNModel, self).__init__()
        self.lstm1 = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.lstm2 = nn.LSTM(hidden_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm1(x)
        out = self.dropout(out)
        out, _ = self.lstm2(out)
        out = self.dropout(out)
        out = self.fc(out[:, -1, :])
        return out
```

```python
input_size = x_train.shape[2]  # Number of features
hidden_size = 128  # Adjust as needed
num_layers = 2
output_size = 1  # Assuming single-value prediction

model = RNNModel(input_size=input_size, hidden_size=100, output_size=1)
criterion = nn.BCEWithLogitsLoss()  # Binary Cross Entropy with Logits
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100  # Adjust as needed
accuracy_list = []

for epoch in range(num_epochs):
    for i, (X_batch, y_batch) in enumerate(dataloader):
        optimizer.zero_grad()
        outputs = model(X_batch.float())
        loss = criterion(outputs.squeeze(), y_batch.float())  # Squeeze to remove the singleton dimension
        loss.backward()
        optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
correct_predictions = 0
total_samples = 0
test_loss =0

with torch.no_grad():
  model.eval()
  for i, (x_test_tensor, y_test_tensor) in enumerate(test_dataloader):

    val_outputs = model(x_test_tensor.float())
    test_loss += criterion(val_outputs.squeeze(), y_test_tensor.float())  # Squeeze to remove the singleton dimension

    # Calculate accuracy
    predicted_labels = (torch.sigmoid(val_outputs) > 0.5).float().squeeze()
    correct_predictions += (predicted_labels == y_test_tensor).sum().item()
    total_samples += len(y_test_tensor)
  accuracy = correct_predictions / total_samples


  print(f'Validation Loss: {test_loss / len(test_dataloader):.4f}, Accuracy: {accuracy:.4f}')
```

## Train CNN Model

```python
class CNNModel(nn.Module):
    def __init__(self, input_size, num_channels, output_size):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=num_channels, kernel_size=3)
        self.relu = nn.ReLU()
        self.fc = nn.Linear(num_channels, output_size)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = x.mean(dim=2)
        x = self.fc(x)
        return x
```

```python

# Hyperparameters
input_size = 10  # Adjust based on your input data size
num_channels = 100
output_size = 1  # Binary classification, so output is 1-dimensional
learning_rate = 0.001
epochs = 10

# Create the model, loss function, and optimizer
model = CNNModel(input_size, num_channels, output_size)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
num_epochs = 100  # Adjust as needed
accuracy_list = []

for epoch in range(num_epochs):
    for i, (X_batch, y_batch) in enumerate(dataloader):
        y_pred = model(X_batch.float())
        loss = criterion(y_pred.squeeze(), y_batch.float())

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()


    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
correct_predictions = 0
total_samples = 0
test_loss =0

with torch.no_grad():
  model.eval()
  for i, (x_test_tensor, y_test_tensor) in enumerate(test_dataloader):

    val_outputs = model(x_test_tensor.float())
    test_loss += criterion(val_outputs.squeeze(), y_test_tensor.float())  # Squeeze to remove the singleton dimension

    # Calculate accuracy
    predicted_labels = (torch.sigmoid(val_outputs) > 0.5).float().squeeze()
    correct_predictions += (predicted_labels == y_test_tensor).sum().item()
    total_samples += len(y_test_tensor)
  accuracy = correct_predictions / total_samples


  print(f'Validation Loss: {test_loss / len(test_dataloader):.4f}, Accuracy: {accuracy:.4f}')
```
